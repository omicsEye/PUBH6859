# To explore the changes in metabolites between RYGB mice at week 1 and month 3 we will do a paired t test between all metabolites across these samples and plot a boxplot for those metabolites which have t-test p-values<0.05 

```python
#  We will iterate through all of the metabolites and do
#  a paired t test for the RYGB samples at 1 week and 3 months
for x in range(3,594): 
  thisMetabolite=df_join.iloc[:,x]
  values=np.asarray(thisMetabolite, dtype=np.float64, order='C')
  res=stats.ttest_rel(values[0:4],values[15:19]) #Conduct paired t-test
  if(res[1]<0.05): #If result is significant
    print(res) #Print t statistic and p-value
    print(df_join.columns[x]) #Print name of metabolite
    #Compose dataframe with the values we want
    all_data = pd.DataFrame({'Week 1 RYGB':values[0:4],
                             'Month 3 RYGB':values[15:19]})
    print(all_data) #print values
    # rectangular box plot
    bplot1=all_data.boxplot()
    bplot1.plot()
    plt.show()
```

# Conclusion
### The metabolite that is most statistically different between week 1 and month 3 of the RYGB mice is 3-methyl-2-oxobutyrate but overall, there are many metabolites that differ.
