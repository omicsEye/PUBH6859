# Question
My question is: Given the Metabolite statistics, how to predict the health outcome using Random Forest under different Time?

```python
# split the data into week 1 and month 3
week_1 = df_join.loc[:14, :]
week_1_X = week_1.iloc[:, 3:]
week_1_y = week_1.iloc[:, 1]

month_3 = df_join.loc[15:, :]
month_3_X = month_3.iloc[:, 3:]
month_3_y = month_3.iloc[:, 1]
```

```python
from sklearn.tree import DecisionTreeClassifier

# make a decision tree and train with week 1 data
tree_w1 = DecisionTreeClassifier()
tree_w1.fit(week_1_X, week_1_y)

print(f'Accuracy for week 1: {tree_w1.score(week_1_X, week_1_y)}')
```

```python
from sklearn import tree
# visualization
tree.plot_tree(tree_w1)
```
